```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data_iris = load_iris()
x_test, x_value, y_test, y_value = train_test_split(data_iris.data[:, 2:4], data_iris.target, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
model = GaussianNB().fit(x_test, y_test)
nb_predictions = model.predict(x_value)
```

Проверка эффективности классификации

```python
accuracy_score(y_value, nb_predictions)
```
