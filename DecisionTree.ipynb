```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn import tree
import graphviz
from sklearn.model_selection import train_test_split
```

```python
data_iris = load_iris()
x_train, x_value, y_train, y_value = train_test_split(data_iris.data[:, 2:4], data_iris.target, random_state = 0)
des_tree = DecisionTreeClassifier()
model_tree = des_tree.fit(x_train, y_train)
```

```python
def print_graph(source_data):
    dot_data = tree.export_graphviz(source_data, out_file=None, 
                                    feature_names = data_iris.feature_names[2:4],
                                    class_names = data_iris.target_names, filled = False)
    return graphviz.Source(dot_data)
```

```python
print_graph(model_tree)
```
