Разделение датасета на тренировочный и тестовый

```python
from sklearn.model_selection import train_test_split
```

```python
x_test, x_value, y_test, y_value = train_test_split(data.data[:,2:4], data['target'], random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5) #5 ближайших соседей
```

Обучение модели

```python
knn_model = knn.fit(x_test, y_test)
```

```python
knn_pred = knn_model.predict(x_value)
```

Проверка, на сколько точным была классификация

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_value, knn_pred)
```
